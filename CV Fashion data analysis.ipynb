```python
diction={'img_id':[],'source':[],'pair_id':[], 'item_no':[],'segmentation':[],'scale':[],
         'viewpoint':[],'zoom_in':[],'landmarks':[],'style':[],'bounding_box':[],
         'category_id':[],'occlusion':[],'category_name':[]
        }
def FileReader(path):        
        with open(path) as file:
            content = json.load(file)
            if int(Path(path).stem)%2000 == 0:
                print('Processing image: ',Path(path).stem)
            for i in range(len(content.keys())-2):
                diction['img_id'].append(Path(path).stem)
                diction['source'].append(content['source'])
                diction['pair_id'].append(content['pair_id'])
                diction['item_no'].append(i+1)
                item = 'item'+str(i+1)
                diction['segmentation'].append(content[item]['segmentation'])
                diction['scale'].append(content[item]['scale'])
                diction['viewpoint'].append(content[item]['viewpoint'])
                diction['zoom_in'].append(content[item]['zoom_in'])
                diction['landmarks'].append(content[item]['landmarks'])
                diction['style'].append(content[item]['style'])
                diction['bounding_box'].append(content[item]['bounding_box'])
                diction['category_id'].append(content[item]['category_id'])
                diction['occlusion'].append(content[item]['occlusion'])
                diction['category_name'].append(content[item]['category_name'])

           
```

## Processing training data to convert it to YOLO format
    This involves -
    1. Reading the image files and getting its dimensions
    2. Converting bounding box parameters from COCO format to YOLO format
    3. Writing the category and bounding box attributes for each image into a corresponding text file
    4. Creating required txt files as input to darknet

```python
print('Total training data: ',train_data.img_id.nunique())
print('Size of training data: ', train_data.shape)
```

```python
print('Total validation data: ',val_data.img_id.nunique())
print('Size of validation data: ', val_data.shape)
```

```python
'''This function uses the given bounding box attributes in coco format (x1,y1,x2,y2)
to obtain bounding box center and size(width,height)
'''
def Yolo_bounding_box(bb_list):
    x1,y1,x2,y2 = bb_list
    w = x2-x1
    h = y2-y1
    x_c = x1+(w/2)
    y_c = y1+(h/2)
    return [x_c,y_c,w,h]
    
    
```

```python
train_data['YOLO_bbox']=train_data['bounding_box'].apply(Yolo_bounding_box)
```

```python
train_data.head()
```

```python
img_path = glob.glob(f'{rootpath}\\train\\train\\image\\*.jpg')
```

```python
""" This function reads each image and obtains its shape """

img_dict = {'img_id':[], 'x_h':[],'y_w':[]}
def read_image(img_path):
    for path in img_path:
        im = cv2.imread(path)
        x_h,y_w,_ = im.shape
        img_dict['img_id'].append(Path(path).stem)
        img_dict['x_h'].append(x_h)
        img_dict['y_w'].append(y_w)
        
```

```python
read_image(img_path)
```

```python
img_data = pd.DataFrame.from_dict(img_dict)
```

```python
img_data.to_csv("train_img_shape.csv")
```

```python
final_data = pd.merge(train_data,img_data,on='img_id')       
    
```

```python
final_data = final_data.drop(['source', 'pair_id', 'item_no', 'segmentation', 'scale',
       'viewpoint', 'zoom_in', 'landmarks', 'style', 'bounding_box', 'occlusion'],axis=1)
```

```python
final_data.head()
```

```python
"""Here we convert each input image bounding box to YOLO format by normalizing the previously
calculated bounding box dimensions with the corresponding width and height"""

def convert_to_yolo(bb,x_h,y_w):
    x_c,y_c,w,h = bb
    return [x_c/y_w,y_c/x_h,w/y_w,h/x_h]

print(convert_to_yolo([233.0, 325.5, 466, 593],624,468))
```

```python
final_data['yolo_format'] = final_data.apply(lambda row : convert_to_yolo(row['YOLO_bbox'],row['x_h'],row['y_w']),axis=1) 
```

```python
final_data.head()
```

```python
final_data.to_csv('Final_Yolo_train.csv')
```

```python
'''Finally, for each image we create a .txt file that contains the attributes for each bounding
in the format -
category_id bounding_box_center_x bounding_box_center_y bounding_box_width bounding_box_height

Note that if the image has more than one categories, all categories are in the same txt file and are line seperated'''

def write_to_folder(data):
    img_ids = (list(set(data['img_id'])))
    print(img_ids)
    
    for id in img_ids:
        filename = r'C:\Users\roshn\Documents\DataAnalytics\Springboard\CV_Fashion_dataset\train\train\YOLO\\'+str(id)+'.txt'
        out = open(filename,'w')
        this_data = data[data['img_id']==id]
        for row in this_data.itertuples():
            out.write(str(int(row.category_id)-1))
            x_c,y_c,w,h = row.yolo_format
            out.write(' ')
            out.write(str(x_c))
            out.write(' ')
            out.write(str(y_c))
            out.write(' ')
            out.write(str(w))
            out.write(' ')
            out.write(str(h))
            out.write('\n')
        out.close()
            
    
```

```python
write_to_folder(final_data)
```
